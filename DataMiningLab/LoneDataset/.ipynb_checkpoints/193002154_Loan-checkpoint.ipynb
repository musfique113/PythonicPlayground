```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
loan_df=pd.read_csv('loan_data_set.csv')
```

# Line plot

```python
sns.lineplot(x='ApplicantIncome', y='LoanAmount', data=loan_df)
plt.title('Loan Amount vs. Applicant Income')
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
plt.show()
```

- Line plot: Shows the relationship between applicant income and loan amount.

# Scatter plot

```python
sns.scatterplot(x='CoapplicantIncome', y='LoanAmount', data=loan_df)
plt.title('Loan Amount vs. Co-applicant Income')
plt.xlabel('Co-applicant Income')
plt.ylabel('Loan Amount')
plt.show()
```

- Scatter plot: Shows the relationship between co-applicant income and loan amount.

# Bar plot

```python
sns.barplot(x='Gender', y='LoanAmount', data=loan_df)
plt.title('Loan Amount by Gender')
plt.xlabel('Gender')
plt.ylabel('Loan Amount')
plt.show()
```

- Bar plot: Compares the loan amount by gender.

# Histogram

```python
sns.histplot(x='LoanAmount', data=loan_df)
plt.title('Loan Amount Distribution')
plt.xlabel('Loan Amount')
plt.show()
```

- Histogram: Displays the distribution of loan amounts.

# Box plot

```python
sns.boxplot(x='LoanAmount', y='Education', data=loan_df)
plt.title('Loan Amount by Education Level')
plt.xlabel('Loan Amount')
plt.ylabel('Education Level')
plt.show()
```

- Box plot: Shows the distribution of loan amounts by education level.

# Area plot

```python
sns.lineplot(x='ApplicantIncome', y='LoanAmount', data=loan_df)
plt.fill_between(loan_df['ApplicantIncome'], loan_df['LoanAmount'], color='skyblue', alpha=0.2)
plt.title('Loan Amount vs. Applicant Income')
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
plt.show()
```

- Area plot: Shows the relationship between applicant income and loan amount with an area filled under the curve.

# Pie chart

```python
loan_status = loan_df['Loan_Status'].value_counts()
plt.pie(loan_status, labels=loan_status.index, autopct='%1.1f%%')
plt.title('Loan Status Distribution')
plt.show()
```

- Pie chart: Shows the distribution of loan statuses.

# Hexbin plot

```python
sns.jointplot(x='ApplicantIncome', y='LoanAmount', kind='hex', data=loan_df)
plt.title('Loan Amount vs. Applicant Income')
plt.show()
```

- Hexbin plot: Shows the relationship between applicant

# Violin plot

```python
sns.violinplot(x='Credit_History', y='LoanAmount', hue='Education', data=loan_df, split=True)
plt.title('Loan Amount by Credit History and Education Level')
plt.xlabel('Credit History')
plt.ylabel('Loan Amount')
plt.show()
```

- Violin plot: Shows the distribution of loan amounts by credit history and education level

# Pair plot

- Pair plot: Displays pairwise relationships

```python
sns.pairplot(loan_df, hue='Loan_Status')
plt.show
```

# Count plot of loan status

```python
sns.countplot(x='Loan_Status', data=loan_df)
plt.title('Loan Status')
plt.show()
```

# Distribution plot of applicant income

```python
sns.displot(x='ApplicantIncome', data=loan_df)
plt.title('Applicant Income')
plt.show()
```

# Bivariate and Multivariate Analysis

```python
sns.pairplot(data=loan_df, hue='Loan_Status', diag_kind='kde')
plt.title('Pair Plot')
plt.show()
```

```python
sns.scatterplot(x='ApplicantIncome', y='LoanAmount', data=loan_df)
plt.title('Loan Amount vs. Applicant Income')
plt.xlabel('Applicant Income')
plt.ylabel('Loan Amount')
plt.show()
```
