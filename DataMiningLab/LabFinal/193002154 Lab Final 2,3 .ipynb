```python
import pandas as pd

df = pd.read_csv("/content/B.csv")

print(df.isnull().sum())

```

```python

df_clean = df.dropna()
print(df_clean.isnull().sum())
```

```python
# Find highest, average, and lowest values of Glucose
glucose_max = df['Glucose'].max()
glucose_avg = df['Glucose'].mean()
glucose_min = df['Glucose'].min()

print("Highest Glucose:", glucose_max)
print("Average Glucose:", glucose_avg)
print("Lowest Glucose:", glucose_min)
```

```python
# Find Age values for highest, average, and lowest Glucose
age_highest_glucose = df.loc[df['Glucose'] == glucose_max, 'Age']
age_average_glucose = df.loc[df['Glucose'] == glucose_avg, 'Age']
age_lowest_glucose = df.loc[df['Glucose'] == glucose_min, 'Age']

print("Age for highest Glucose:", age_highest_glucose.values)
print("Age for average Glucose:", age_average_glucose.values)
print("Age for lowest Glucose:", age_lowest_glucose.values)
```

```python
# Find the number of entries above and below average Glucose
above_average = df[df['Glucose'] > glucose_avg].shape[0]
below_average = df[df['Glucose'] < glucose_avg].shape[0]

print("Entries above average Glucose:", above_average)
print("Entries below average Glucose:", below_average)
```
